```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")

read_dataset = pd.read_csv("/kaggle/input/export-dataset/garment_tea_rubber_and coconut_exports.csv")
read_dataset.head(10)
```

```python
read_dataset.head(5)
```

```python
# overall tea, rubber, garment exports

exports = read_dataset.head(3)
exports
```

```python
coconut_by_country = read_dataset.iloc[7:].reset_index(drop=True)
coconut_by_country.head(10)
```

```python
column_sums = coconut_by_country.sum().to_frame().T
column_sums.index = ['Total']
df_final = pd.concat([coconut_by_country, column_sums], ignore_index=False)
df_final
```

```python
df_final[-1:]
```

```python
new_df = pd.concat([exports, df_final[-1:]], ignore_index=False)
new_df
```

```python
new_df_reset = new_df.reset_index(drop=True)
new_df_reset.loc[3, 'Product'] = 'Coconut Exports (US $ mn)'
new_df_reset
```

```python
sns.boxplot(data=new_df_reset)
```

```python
coconut_by_country2 = coconut_by_country
coconut_by_country2['PercentageChange'] = ((coconut_by_country.iloc[:, -1] - coconut_by_country.iloc[:, 1]) / coconut_by_country.iloc[:, 1]) * 100
coconut_by_country2
```

```python
coconut_by_country2 = coconut_by_country2.sort_values(by='PercentageChange', ascending=False)
coconut_by_country2
```

```python
coconut_by_country.describe()
```

```python
subset_df = coconut_by_country.head(5).reset_index(drop=True)

subset_df.set_index('Product', inplace=True)
df_transposed = subset_df.transpose()
```

```python

sns.set(style="whitegrid")  
plt.figure(figsize=(10, 6))
           
for country in df_transposed.columns:
    sns.lineplot(x=df_transposed.index, y=df_transposed[country], label=country,linewidth=5, alpha=0.8)

plt.title('Coconut Exports by Country per Year (US $ mn)',fontweight='bold')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('US $ mn',fontweight='bold')
plt.legend(title='Countries')
plt.xticks(rotation=0)  
plt.tight_layout() 

plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')

plt.show()
```

```python
df_melted = exports.melt(id_vars=['Product'], var_name='Year', value_name='Value')
df_melted['Product']  = df_melted['Product'].str.replace('Exports (US $ mn)', '')
```

```python
plt.figure(figsize=(10, 6))  
sns.barplot(data=df_melted, x='Year', y='Value', hue='Product')

plt.title('Exports Over Years (US $ mn)')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('US $ mn',fontweight='bold')
plt.xticks(rotation=0)  
plt.tight_layout() 
plt.legend(title='Product')
plt.show()
```

```python
subset_exports = exports.reset_index(drop=True)
subset_exports['Product']  = subset_exports['Product'].str.replace('Exports (US $ mn)', '')
subset_exports.set_index('Product', inplace=True)
exports_transposed = subset_exports.transpose()
exports_transposed
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  
plt.figure(figsize=(10, 6))
           
for product in exports_transposed.columns:
    sns.lineplot(x=exports_transposed.index, y=exports_transposed[product], label=product,linewidth=5, alpha=0.8)

plt.title('Coconut Exports by Country (US $ mn)',fontweight='bold')
plt.xlabel('Year',fontweight='bold')
plt.ylabel('Value',fontweight='bold')
plt.legend(title='Product')
plt.xticks(rotation=0)  
plt.tight_layout() 

plt.gca().spines['top'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['left'].set_color('black')

plt.show()
```
